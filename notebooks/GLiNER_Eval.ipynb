```python
# available models: https://huggingface.co/urchade

model = GLiNER.from_pretrained("urchade/gliner_small-v2.1")
model.eval()
print("ok")
```

```python
text = """
Libretto by Marius Petipa, based on the 1822 novella ``Trilby, ou Le Lutin d'Argail`` by Charles Nodier, first presented by the Ballet of the Moscow Imperial Bolshoi Theatre on January 25/February 6 (Julian/Gregorian calendar dates), 1870, in Moscow with Polina Karpakova as Trilby and Ludiia Geiten as Miranda and restaged by Petipa for the Imperial Ballet at the Imperial Bolshoi Kamenny Theatre on January 17–29, 1871 in St. Petersburg with Adèle Grantzow as Trilby and Lev Ivanov as Count Leopold.
"""

labels = ["person", "book", "location", "date", "actor", "character"]

entities = model.predict_entities(text, labels, threshold=0.4)

for entity in entities:
    print(entity["text"], "=>", entity["label"])
```

```python
import json
from collections import defaultdict

def calculate_metrics(json_file):
    """
    Calculate precision, recall, and F1-score from a JSON file with nested lists, using explicit matching via loops.

    Args:
        json_file (str): Path to the JSON file.

    Returns:
        dict: Metrics including overall precision, recall, F1-score, and per-category metrics.
    """
    metrics = {"overall": {"TP": 0, "FP": 0, "FN": 0}}
    per_category_metrics = defaultdict(lambda: {"TP": 0, "FP": 0, "FN": 0})

    # Load JSON data
    with open(json_file, "r") as f:
        data = json.load(f)

    # Iterate through each item in the nested lists
    for trues, preds in zip(data["all_trues"], data["all_preds"]):
        matched_trues = set()
        matched_preds = set()

        # Try to match each prediction with any ground truth
        for pred in preds:
            pred_tuple = tuple(pred)
            found_match = False
            for idx, true in enumerate(trues):
                # true_tuple = tuple(true)
                start, end, label = true
                start_pred, end_pred, label_pred, _ = pred
                if start == start_pred and end == end_pred == end and label == label_pred:
                    metrics["overall"]["TP"] += 1
                    per_category_metrics[label]["TP"] += 1
                    matched_trues.add(idx)
                    found_match = True
                    break

            if not found_match:  # If no match is found, it's a False Positive
                metrics["overall"]["FP"] += 1
                per_category_metrics[label_pred]["FP"] += 1
                matched_preds.add(pred_tuple)

        # Remaining ground truths that were not matched are False Negatives
        for idx, true in enumerate(trues):
            if idx not in matched_trues:
                true_tuple = tuple(true)
                metrics["overall"]["FN"] += 1
                per_category_metrics[true_tuple[2]]["FN"] += 1
        # print("matched_preds", matched_preds)
        # break

    # Calculate precision, recall, and F1
    def compute_scores(tp, fp, fn):
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0.0
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0
        return precision, recall, f1
    print(metrics)
    overall_precision, overall_recall, overall_f1 = compute_scores(
        metrics["overall"]["TP"], metrics["overall"]["FP"], metrics["overall"]["FN"]
    )

    per_category_scores = {}
    for category, counts in per_category_metrics.items():
        precision, recall, f1 = compute_scores(counts["TP"], counts["FP"], counts["FN"])
        per_category_scores[category] = {"precision": precision, "recall": recall, "f1": f1}

    return {
        "overall": {"precision": overall_precision, "recall": overall_recall, "f1": overall_f1},
        "per_category": per_category_scores,
    }
# Example Usage
json_file = "/content/GLiNER/results.json"  # Path to your JSON file
metrics = calculate_metrics(json_file)

print("Overall Metrics:")
print(metrics["overall"])
print("\nPer-Category Metrics:")
for category, scores in metrics["per_category"].items():
    print(f"{category}: {scores}")
```
